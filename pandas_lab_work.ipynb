```python
#8.joining dataframe
student_detail={'student_id':['s1','s2','s3','s4','s5'],
                'name':['Danniella Fenton','Ryder Storey','Bryce Jensen','Ed Bernal','Kwame Morin'],
                'marks':[200,210,190,222,199]}

student_data2={'student_id':['s4','s5','s6','s7','s8'],
               'name':['Scarlette Fisher','Carla Williamson','Dante Morse','Kaiser William','Madeeha preston'],
               'marks':[201,200,198,219,201]}
student_detail=pd.DataFrame(student_detail)
student_data2=pd.DataFrame(student_data2)
pd.concat([student_detail,student_data2],axis=0)
```
