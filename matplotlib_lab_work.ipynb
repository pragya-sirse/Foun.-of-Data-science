```python
#2.line chart of financial data
import csv
financial_data=[["Date","Open","High","Low","Close"],
               ["10-03-16",774.25,776.065003,769.5,772.55998],
               ["10-04-16",776.030029,778.710022,772.890015,776.429993],
               ["10-05-16",779.309998,782.070007,775.650024,776.469971],
               ["10-06-16",779,780.47998,775.539978,776.859985],
               ["10-07-16",779.659973,779.659973,770.75,775.080017]]
with open('fdata.csv','w',newline='') as csvfile:
  writer=csv.writer(csvfile)
  writer.writerows(financial_data)
```

```python
df=pd.read_csv('fdata.csv')
df
```

```python
df['Date']=pd.to_datetime(df['Date'],format='%m-%d-%y')
df.set_index('Date',inplace=True)
plt.figure(figsize=(10,5))
plt.plot(df.index,df['Close'],linestyle='-')
plt.title('Alphabet Inc.b/w oct 3-2016 to oct 7-2016')
plt.xlabel('Date')
plt.ylabel('Close')
plt.show()
```

```python
#3.
import matplotlib.pyplot as plt

# Data for the first line
x1 = [10, 15, 20, 25, 30]
y1 = [25, 30, 40, 45, 50]

# Data for the second line
#x2 = [10, 15, 20, 25, 30]
#y2 = [20, 25, 30, 35, 40]

# Plot the first line with a solid line style
plt.plot(x1,  linestyle='-', label='Line 1')

# Plot the second line with a dashed line style
plt.plot(y1,  linestyle='--', label='Line 2')

# Set labels for x-axis, y-axis, and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Two Lines with Different Styles')

# Add legend
plt.legend()

# Display the plot
plt.show()
```
