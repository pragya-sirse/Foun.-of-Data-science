```python
# from a pandas dataframe
batsman=pd.read_csv('/content/sharma-kohli.csv')
batsman
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
#plottng multiple graph
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
# label title
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.title('Rohit Sharma vs kohli Carrer Comparison')
plt.xlabel('season')
plt.ylabel('Runs')
```

```python
#colors(hex) and line(width and style) and marker(size)
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F')
plt.plot(batsman['index'],batsman['RG Sharma'],color='black')
plt.title('Rohit Sharma vs kohli Carrer Comparison')
plt.xlabel('season')
plt.ylabel('Runs')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F',linestyle='dashed',linewidth=3)
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',linestyle='dotted',linewidth=2)
plt.title('Rohit Sharma vs kohli Carrer Comparison')
plt.xlabel('season')
plt.ylabel('Runs')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F',linestyle='dashed',linewidth=3,marker='D',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',linestyle='dotted',linewidth=2,marker='+')
plt.title('Rohit Sharma vs kohli Carrer Comparison')
plt.xlabel('season')
plt.ylabel('Runs')
```

```python
#legend ->location
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F',linestyle='dashed',linewidth=3,marker='D',markersize=10,label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',linestyle='dotted',linewidth=2,marker='+',label='Rohit')
plt.title('Rohit Sharma vs kohli Carrer Comparison')
plt.xlabel('season')
plt.ylabel('Runs')
plt.legend()
```

```python
#limiting axes
price=[48000,54000,60000,90000,32000,900000]
year=[2015,2016,2017,2018,2019,2020]
plt.plot(year,price)
plt.ylim(0,90000)
plt.xlim(2017,2019)
```

```python
# multiple bar chart
df=pd.read_csv('/content/batsman_season_record.csv')
df
```

```python
plt.bar(np.arange(df.shape[0])-0.2,df['2015'],width=0.2)
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2)
plt.bar(np.arange(df.shape[0])+0.2,df['2017'],width=0.2)
plt.xticks(np.arange(df.shape[0]),df['batsman'])
plt.show()
```

```python
#problem
childern=[10,20,30,40,50]
colors=['red','red','red','red','blue','green','blue','green','yellow','pink']
plt.bar(colors,childern,color='black')
plt.xsticks(rotation='vertical')
```

## Stacked bar chart

```python
df
```

```python
# stacked bar chart
plt.bar(df['batsman'],df['2015'],label='2015')
plt.bar(df['batsman'],df['2016'],bottom=df['2015'],label='2016')
plt.bar(df['batsman'],df['2017'],bottom=df['2016']+df['2015'],label='2017')
plt.legend()
```
